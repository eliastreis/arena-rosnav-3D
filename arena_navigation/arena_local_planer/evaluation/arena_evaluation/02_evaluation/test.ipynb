```python
import numpy as np
import seaborn as sns
import pandas as pd
data = pd.DataFrame()
a = [1,2,1,2,5,1,2]
b = [5,2,5,1,2,5,1]
c = [5,1,2,5,2,5,1]
d = [1,5,1,2,8,2,2]
e = [1,2,5,2,6,3,4]
f = [5,1,2,5,3,1,2]
g = [1,2,5,2,3,4,2]
h = [1,5,2,3,1,2,4]
i = [4,2,1,2,4,2,1]
j = [7,2,3,5,2,5,1]
k = [5,2,5,2,3,1,1]
data = pd.concat([data,pd.DataFrame(list(zip(["a"]*len(a),a)), columns = ["index","numbers"])])
data = pd.concat([data,pd.DataFrame(list(zip(["b"]*len(b),b)), columns = ["index","numbers"])])
data = pd.concat([data,pd.DataFrame(list(zip(["c"]*len(c),c)), columns = ["index","numbers"])])
data = pd.concat([data,pd.DataFrame(list(zip(["d"]*len(d),d)), columns = ["index","numbers"])])
data = pd.concat([data,pd.DataFrame(list(zip(["e"]*len(e),e)), columns = ["index","numbers"])])
data = pd.concat([data,pd.DataFrame(list(zip(["f"]*len(f),f)), columns = ["index","numbers"])])
data = pd.concat([data,pd.DataFrame(list(zip(["g"]*len(g),g)), columns = ["index","numbers"])])
data = pd.concat([data,pd.DataFrame(list(zip(["h"]*len(h),h)), columns = ["index","numbers"])])
data = pd.concat([data,pd.DataFrame(list(zip(["i"]*len(i),i)), columns = ["index","numbers"])])
data = pd.concat([data,pd.DataFrame(list(zip(["j"]*len(j),j)), columns = ["index","numbers"])])
data = pd.concat([data,pd.DataFrame(list(zip(["k"]*len(k),k)), columns = ["index","numbers"])])
sns.violinplot(x="index",y="numbers",data=data, inner="point")
```
